```python
import os
import glob
```

```python
# 파일 이름 변경 규칙
def rename_file(file_name):
    mapping = {
        "cifar100": "1.cifar100.yaml",
        "imagenet_r": "2.imagenet_r.yaml",
        "cars196": "3.cars196.yaml",
        "cub200": "4.cub200.yaml",
        "eurosat": "5.eurosat.yaml",
        "resisc45": "6.resisc45.yaml",
        "chestx": "7.chestx.yaml",
        "isic": "8.isic.yaml",
        "cropdisease": "9.cropdisease.yaml",
    }

    for keyword, new_name in mapping.items():
        if keyword in file_name:
            return new_name
    return None
```

```python
# 최상위 디렉토리 경로 설정
top_directory = "."  # 여기에 최상위 디렉토리 경로를 입력

# 디렉토리 순회하며 파일 이름 변경
for root, dirs, files in os.walk(top_directory):
    if "test_sweep" in root:  # 중간 경로에 test_sweep 포함 여부 확인
        for file_name in files:
            if not file_name.split('.')[0].isdigit():
                new_name = rename_file(file_name)
                if new_name and file_name != new_name:  # 이미 변경된 파일은 제외
                    old_path = os.path.join(root, file_name)
                    new_path = os.path.join(root, new_name)
                    os.rename(old_path, new_path)
                    print(f"Renamed: {old_path} -> {new_path}")

print("File renaming completed.")
```
